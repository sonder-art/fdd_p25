# 03 — Slicing e Indexación (1D y 2D)

- Indexación básica y booleana
- Slicing por rangos y saltos
- `None` para agregar ejes: `n x 1` vs `n x None`
- Diferencias entre 1D y 2D



```python
import numpy as np

arr = np.array([1,2,3,4,5])
idx = arr[[0,2,4]]
mask = (arr % 3 == 1) | (arr % 3 == 2)
(arr[1:], arr[::2], idx, arr[mask])
mask # muestra aquellos valores que cumplen con el criterio del mask
```

```python
# None agrega un eje nuevo
v = np.arange(5)
v = v.reshape(5,1)
col = v[:, None]   # n x 1
row = v[None, :]   # 1 x n
(v, v.shape, col.shape, row.shape, row, col)

```

```python
# 2D slicing
M = np.arange(1, 101).reshape(10,10)
(
    M,
    M[0, :],      # primera fila
    M[:, 0],      # primera columna
    M[1:2, 1:2],   # submatriz  excluye 1 e incluye 2
)
```
