# 02 â€” Arrays de NumPy y DTypes

- Crear arrays: `np.array`, `np.zeros`, `np.ones`, `np.arange`, `np.linspace`
- `dtype` y casting vs tipos nativos de Python
- `shape`, `ndim`, `size`
- Copia vs vista (`copy`, slicing)



```python
import numpy as np

# creacion
arr1 = np.array([1,2,3])
arr2 = np.zeros(( 2,10))
arr3 = np.ones((2, 3))
arr4 = np.arange(0,10,2)
arr5 = np.linspace(0, 1, 5)

arr1.dtype, arr2.size, arr3.ndim, arr4.size, arr5
```

```python
# dtypes y casting
arr = np.array([1, 2, 3], dtype=np.int64)
arr_float = arr.astype(np.float64)
(type(1), isinstance(1, int), isinstance(1, np.integer), arr_float.dtype)
```

```python
# copia vs vista
base = np.arange(5)
vie = base[1:4]
copy = base[1:4].copy
vie[0] = -1
(base, vie, copy)
```
