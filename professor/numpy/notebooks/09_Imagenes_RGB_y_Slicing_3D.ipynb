# 09 — Imágenes: RGB y slicing 3D

Basado en `legacy/00_numpy.ipynb` (sección de imágenes), con más explicación.

- Cargar imagen con `matplotlib.pyplot.imread`
- `shape` y dimensión de canales
- Separar canales R/G/B mediante slicing
- Manipular canales y visualizar resultados



```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# Intentamos usar el asset local de legacy; si no existe, mensaje claro
img_path = Path("181cab12f0d3d4168ddac0edfbd1f157.jpg")
if not img_path.exists():
    raise FileNotFoundError("No se encontró la imagen en legacy/. Copia los assets de legacy o ajusta la ruta.")

img = plt.imread(img_path)
img = np.array(img)
img.shape
```

```python
# Visualización base
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
# Extraer canales
red = img[:, :, 0]
green = img[:, :, 1]
blue = img[:, :, 2]
(red.shape, green.shape, blue.shape)
```

```python
# Manipular canales
img_red = img.copy()
img_red[:, :, 1] = 0
img_red[:, :, 2] = 0

img_green = img.copy()
img_green[:, :, 0] = 0
img_green[:, :, 2] = 0

img_blue = img.copy()
img_blue[:, :, 0] = 0
img_blue[:, :, 1] = 0

fig, axes = plt.subplots(1, 3, figsize=(12,4))
for ax, im, title in zip(axes, [img_red, img_green, img_blue], ["Rojo", "Verde", "Azul"]):
    ax.imshow(im)
    ax.set_title(title)
    ax.axis('off')
plt.show()
```
